```python
notas = pd.read_csv('ratings.csv')
notas.head()
```

```python
notas_media_filme = notas.groupby('movieId').mean()['rating']
notas_media_filme.head()
```

```python
ax = sns.histplot(notas_media_filme.values)
ax.set(xlabel='Nota Media', ylabel='Densidade')
ax.set_title('Media de votos em filmes no MovieLens 100k')
```

```python
from statsmodels.stats.weightstats import ztest

ztest(notas1.rating, value = 3.4320503405352594)
```

```python
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = (temp[0:i].mean())
    statis, p = ztest(temp[0:i],value = 3.4320503405352594)
    return(i,media,p)

valores = np.array([calcula_teste(i) for i in range(1, len(temp))])
#medias = [temp[0:i].mean() for i in range(1, len(temp))]
plt.plot(valores[:,0],valores[:,1])
plt.plot(valores[:,0],valores[:,2])
plt.hlines(y= 0.05, xmin = 2, xmax = len(temp), colors='r')
```
